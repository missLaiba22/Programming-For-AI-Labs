# Lab Task 02

Import the dataset from this address https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user without downloading it manually

```python
# YOUR CODE HERE
csv_url='https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
df=pd.read_csv(csv_url)
print(df)
```

Assign it to a variable called users and use the 'user_id' as index. choose the seperater/delimiter carefully.

```python
# YOUR CODE HERE
users = pd.read_csv(csv_url, index_col='user_id', delimiter='|')

# Print the `users` dataframe
print(users)
```

Display 1/3rd rows

```python
# YOUR CODE HERE
print(users.head(int(len(users)/3)))
```

Display the entire occupation column

```python
# YOUR CODE HERE
print(users['occupation'])
```

How many different occupations are in this dataset?

```python
# YOUR CODE HERE
num_occupations=users['occupation'].nunique()
print(num_occupations)
```

What is the most occuring occupation for each gender seperatly?

```python
# YOUR CODE HERE

mode_occ_by_gender = users.groupby('gender')['occupation'].value_counts().groupby('gender').head(1)
print(mode_occ_by_gender)
```

What is the mean age of users?

```python
# YOUR CODE HERE
mean_age=users['age'].mean()
print(mean_age)
```

Display the users who have age more than 40

```python
# YOUR CODE HERE
users_over_40=users[users['age']>40]
print(users_over_40)
```

Find max and min value from each numeric column

```python
# YOUR CODE HERE
max_values=users.max(axis=0)
min_values=users.min(axis=0)
print("max: ",max_values)
print("min: ",min_values)
```

Find smallest and largest string from occupation column

```python
# YOUR CODE HERE
# Find the smallest string in the 'occupation' column
smallest_str = users['occupation'].sort_values().iloc[0]

# Find the largest string in the 'occupation' column
largest_str = users['occupation'].sort_values(ascending=False).iloc[0]

# Print the smallest and largest string
print(smallest_str)
print(largest_str)
```
